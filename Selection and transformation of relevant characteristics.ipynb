```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
#Creando un Data frame de ejemplo con filas duplicadas

# Creando un Data frame de ejemplo con filas duplicadas
data = {
    'D': ['cat', 'dog', 'cat', 'bird', 'dog', 'cat', 'bird', 'cat', 'dog', 'cat'],
    'E': [1, 1, 2, 2, 3, 3, 4, 4, 5, 1],
    'F': ['a', 'b', 'a', 'b', 'a', 'b', 'a', 'b', 'a', 'a']
}

df= pd.DataFrame(data)
print(df)
```

```python
#Añadir una fila duplicada
first_row = df.iloc[0]

# Añade la fila duplicada al DataFrame
df.loc[len(df)] = first_row
print(df)
```

```python
#Codificación de variable categóricas, conversión de categorías a numeros
encoder= LabelEncoder()
df['D']= encoder.fit_transform(df['D'])
print(df)
```

```python
#Eliminacion de duplicados, eliminar filas duplicadas
df.drop_duplicates(inplace=True)
print(df)
```

```python
df['E']= df['E'].apply(np.log)
print(df)
```
